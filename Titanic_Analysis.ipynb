```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

# Check if it loaded
print("Success! Data loaded.")
df.head()
```

```python
df['age'] = df['age'].fillna(df['age'].median())
```

```python
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Gender')
```

```python
sns.countplot(x='pclass', hue='survived', data=df)
plt.title('Survival Count by Class')
```

While 3rd class had a high number of survivors in total, the survival rate was lowest for this group due to the high total volume of passengers.

```python

cross_tab = pd.crosstab(df['class'], df['survived'], normalize='index')

# Plotting the probability
cross_tab.plot(kind='bar', stacked=True, color=['#e74c3c', '#3498db'], figsize=(8, 5))

plt.title('Proportion of Survival vs Death by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Probability (0.0 to 1.0)')
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])
plt.xticks(rotation=0)
plt.show()
```

Key Findings:

    Gender Gap: Women had a significantly higher survival rate than men across all classes.
    Class Disparity: Passengers in 1st class were prioritized, while 3rd-class passengers had the lowest survival rates.
    Data Cleaning: Missing age values were handled using median imputation to maintain dataset integrity.
